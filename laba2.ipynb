```python
import matplotlib.pyplot as plt
import numpy as np

data = {
    "": [-5, -3, -2, 0, 2, 4, 6],
    "-": [-7, -4, -1, 1, 3, 5, 7],
    "": [-15, -12, -10, -8, -6, -4, -2],
    "": [-10, -8, -6, -4, -2, 0, 2],
    "": [-8, -6, -4, -2, 0, 2, 4]
}

plt.figure(figsize=(12, 6))
days = np.arange(1, 8)
for city, temperatures in data.items():
    plt.plot(days, temperatures, marker='o', label=city)
plt.xlabel('  ')
plt.ylabel(', C')
plt.title('       ')
plt.xticks(days, ['', '', '', '', '', '', ''])
plt.legend()
plt.grid()
plt.show()

cities = list(data.keys())
temperature_array = np.array(list(data.values()))
plt.figure(figsize=(10, 6))
plt.contourf(days, cities, temperature_array, levels=7, cmap='coolwarm_r')
cbar = plt.colorbar()
cbar.set_label(', C')
plt.xlabel(' ')
plt.ylabel('')
plt.title('       ')
plt.xticks(days, ['', '', '', '', '', '', ''])
plt.show()
```
