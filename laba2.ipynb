```python
import matplotlib.pyplot as plt
import numpy as np

data = {
    "Москва": [-5, -3, -2, 0, 2, 4, 6],
    "Санкт-Петербург": [-7, -4, -1, 1, 3, 5, 7],
    "Новосибирск": [-15, -12, -10, -8, -6, -4, -2],
    "Екатеринбург": [-10, -8, -6, -4, -2, 0, 2],
    "Казань": [-8, -6, -4, -2, 0, 2, 4]
}

plt.figure(figsize=(12, 6))
days = np.arange(1, 8)
for city, temperatures in data.items():
    plt.plot(days, temperatures, marker='o', label=city)
plt.xlabel('Номер дня недели')
plt.ylabel('Температура, °C')
plt.title('Изменение температуры воздуха в городах по дням недели')
plt.xticks(days, ['Пн', 'Вт', 'Ср', 'Чт', 'Пт', 'Сб', 'Вс'])
plt.legend()
plt.grid()
plt.show()

cities = list(data.keys())
temperature_array = np.array(list(data.values()))
plt.figure(figsize=(10, 6))
plt.contourf(days, cities, temperature_array, levels=7, cmap='coolwarm_r')
cbar = plt.colorbar()
cbar.set_label('Температура, °C')
plt.xlabel('Дни недели')
plt.ylabel('Города')
plt.title('Распределение температуры воздуха по городам и дням недели')
plt.xticks(days, ['Пн', 'Вт', 'Ср', 'Чт', 'Пт', 'Сб', 'Вс'])
plt.show()
```
