```python
import numpy as np
import matplotlib.pyplot as plt

a = np.linspace(1, 20, 5)
b = np.cos(a / 20)
c = np.random.random(20) * 10

y = [4, 4, 1, 5, -1]
s = ['one', 'two', 'three ', 'four', 'five']
```

1) Построить график гистограммы (hist) на основании **массива c**. Кол-во бинов указать = 10. Сделать график грасного цвета (red).

Сделайте **подпись** над графиком выше - Histogramm

```python
fig = plt.figure()
ax = fig.subplots(1)
ax.hist(x=c, bins=10, color='red')
ax.set(title='Histogramm')
plt.show()
```

2) Построить столбчатую диаграмму (barplot) на основании **массивов s и y** (где по оси **OX - s, по OY - y**). Сделать график серого цвета (grey).

Сделайте **подпись** над графиком выше - Bar chart

```python
fig = plt.figure()
ax = fig.subplots(1)
ax.bar(s, y, color='grey')
ax.set(title='Bar chart')
ax.axhline(y=0, color='black', linestyle='-')
plt.show()
```

3) Построить круговую диаграмму (pie) на основании **массивов a и s** (где по x=a, labels=s). **Используйте обязательно параметр explode**, чтобы выдвинуть **первый элемент** на расстояние 0.2, все остальные оставить на своих местах.

Сделайте **подпись** над графиком выше - Pie chart

```python
fig = plt.figure()
ax = fig.subplots(1)
ax.pie(x=a, labels=s, explode=(0.2, 0, 0, 0, 0))
ax.set(title='Pie chart')
plt.show()
```

4) Построить график ящика с усами (boxplot) на основании **массивов a и c** (то есть на самом графике у вас должно быть два ящика!). 

Сделайте **подпись** над графиком выше - Boxplot

```python
fig = plt.figure()
ax = fig.subplots(1)
ax.boxplot([a, c], vert=0)
ax.set(title='Boxplot')
plt.show()
```

5) Построить график панели ошибок (errorbar) на основании **массивов a и b** (где по оси OX - a, по OY - b). В качестве **диапозона ошибок** возьмите следующие значения xerr=1, yerr=0.2

Сделайте **подпись** над графиком выше - Errorbar

```python
fig = plt.figure()
ax = fig.subplots(1)
ax.errorbar(a, b, xerr=1, yerr=0.2)
plt.show()
```

**Например:**

Лучше использовать **plt.subplots()** для данных целей. Если у вас блоки расположены в разных строках, тогда постараться не делать мелкую ширину у графиков, минимум ширина должна быть в половину ширины экрана

```python
# можете использовать следующие значения
# либо выбрать свои данные, а также функции (cos/sin и т д)
N = 100

x = np.arange(N)

y_1 = np.sin(x / 10)
y_2 = np.sin(x / 5)
```

```python
fig = plt.figure()
ax = fig.subplots(1)
ax.plot(x, y_1, '-.', alpha=0.8, color='red')
ax.plot(x, y_2, '-', alpha=0.8, color='blue')
ax.set(xlabel='x', ylabel='y', )
ax.grid(True)
plt.show()
```

```python
fig = plt.figure()
ax = fig.subplots(1, 2)
ax[0].plot(x, y_1, '-.', alpha=0.8, color='red')
ax[1].plot(x, y_2, '-', alpha=0.8, color='blue')
ax[0].set(xlabel='x', ylabel='y')
ax[1].set(xlabel='x')
ax[0].grid(True)
ax[1].grid(True)
plt.show()
```

```python
fig = plt.figure()
ax = fig.subplots(2, 1)
ax[0].plot(x, y_1, '-.', alpha=0.8, color='red')
ax[1].plot(x, y_2, '-', alpha=0.8, color='blue')
ax[0].set(xlabel='x', ylabel='y')
ax[1].set(xlabel='x', ylabel='y')
ax[0].grid(True)
ax[1].grid(True)
plt.show()
```

Подсказка для поиска координаты текста над графиком:

min_ylim, max_ylim = plt.ylim()

```python
fig = plt.figure(figsize=(12, 6))
ax = fig.subplots(1)
mu_list = [-1, 1, 3]
sigma_list = [0.5, 0.65, 0.8]
color_list = ['blue', 'red', 'green']
coords_list = [0.15, 0.35, 0.55]

for mu, sigma, color, coord in zip(mu_list, sigma_list, color_list, coords_list):
    s = np.random.normal(mu, sigma, 5000)
    ax.hist(s, bins=50, alpha=0.5, density=False, color=color)
    ax.axvline(mu, color=color, alpha=0.5)
    ax.text(coord, 0.9, f'Mean y: {mu:.2f}', transform=ax.transAxes, color=color)

ax.set(xticks=np.arange(-4, 7, 1))
plt.show()
```
