```python
import numpy as np
import matplotlib.pyplot as plt
import random

data = { 
"": [1200, 1500, 1800], 
"-": [800, 900, 1000], 
"": [600, 700, 800], 
"": [400, 500, 600], 
"": [300, 400, 500] 
}

plt.figure(figsize=(10, 6))
cities = list(data.keys())
totalsales = [sum(sales) for sales in data.values()]
bars = plt.bar(cities, totalsales, color=['red', 'orange', 'yellow', 'green', 'blue'])

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 50, yval, ha='center', va='bottom')

plt.xlabel('')
plt.title('       ')
plt.show()

plt.figure(figsize=(6, 6))
colors = 'red', 'orange', 'yellow', 'green', 'blue'
plt.pie(totalsales, labels=cities, colors=colors, autopct='%1.1f%%')
plt.title('         ')
plt.show()

plt.figure(figsize=(10, 6))
months = [1, 2, 3]
for city, sales in data.items():
    plt.plot(months, sales, marker='o', label=city)
plt.xlabel(' ')
plt.ylabel(' ')
plt.title('       ')
plt.legend()
plt.show()

orders = [random.randint(1, 5) for _ in range(1000)]
plt.figure(figsize=(8, 6))
plt.hist(orders, bins=[0.5, 1.5, 2.5, 3.5, 4.5, 5.5], color='grey', edgecolor='black')
plt.xlabel('   ')
plt.ylabel(' ')
plt.title('        ')

plt.show()
```
