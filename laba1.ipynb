```python
import numpy as np
import matplotlib.pyplot as plt
import random

data = { 
"Москва": [1200, 1500, 1800], 
"Санкт-Петербург": [800, 900, 1000], 
"Новосибирск": [600, 700, 800], 
"Екатеринбург": [400, 500, 600], 
"Казань": [300, 400, 500] 
}

plt.figure(figsize=(10, 6))
cities = list(data.keys())
totalsales = [sum(sales) for sales in data.values()]
bars = plt.bar(cities, totalsales, color=['red', 'orange', 'yellow', 'green', 'blue'])

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 50, yval, ha='center', va='bottom')

plt.xlabel('Города')
plt.title('Общее количество проданных порций мороженого в каждом городе')
plt.show()

plt.figure(figsize=(6, 6))
colors = 'red', 'orange', 'yellow', 'green', 'blue'
plt.pie(totalsales, labels=cities, colors=colors, autopct='%1.1f%%')
plt.title('Доля продаж мороженого в каждом городе от общего объема продаж')
plt.show()

plt.figure(figsize=(10, 6))
months = [1, 2, 3]
for city, sales in data.items():
    plt.plot(months, sales, marker='o', label=city)
plt.xlabel('Номер месяца')
plt.ylabel('Количество порций')
plt.title('Динамика продаж мороженого в каждом городе по месяцам')
plt.legend()
plt.show()

orders = [random.randint(1, 5) for _ in range(1000)]
plt.figure(figsize=(8, 6))
plt.hist(orders, bins=[0.5, 1.5, 2.5, 3.5, 4.5, 5.5], color='grey', edgecolor='black')
plt.xlabel('Количество порций в заказе')
plt.ylabel('Количество заказов')
plt.title('Распределение продаж мороженого по количеству порций в одном заказе')

plt.show()
```
